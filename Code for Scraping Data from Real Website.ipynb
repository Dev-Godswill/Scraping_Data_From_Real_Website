# Scraping Data from a Real Website (Code)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

# Send a GET request to the URL and retrieve the HTML content
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find and print the HTML content of the entire page
print(soup)

# Find the first and second tables on the page
table_largest_companies = soup.find_all('table')[1]
table_profitable_companies = soup.find_all('table')[2]

# Print the HTML content of the first and second tables
print(table_largest_companies)
print(table_profitable_companies)

# Find all the header cells (th) in the first table
header_cells_largest_companies = table_largest_companies.find_all('th')

# Print the header titles of the first table
for header_cell in header_cells_largest_companies:
    print(header_cell.text.strip())

# Find all the header cells (th) in the second table
header_cells_profitable_companies = table_profitable_companies.find_all('th')

# Print the header titles of the second table
for header_cell in header_cells_profitable_companies:
    print(header_cell.text.strip())

# Extract the header titles and remove the '\n' characters in the first table
table_titles_largest_companies = [header_cell.text.strip() for header_cell in header_cells_largest_companies]
print(table_titles_largest_companies)

# Extract the header titles and remove the '\n' characters in the second table
table_titles_profitable_companies = [header_cell.text.strip() for header_cell in header_cells_profitable_companies]
print(table_titles_profitable_companies)

# Create empty DataFrames with the extracted header titles
df_largest_companies = pd.DataFrame(columns=table_titles_largest_companies)
df_profitable_companies = pd.DataFrame(columns=table_titles_profitable_companies)

# Extract the data rows from the first table
data_rows_largest_companies = table_largest_companies.find_all('tr')[1:]
for row in data_rows_largest_companies:
    row_data_largest_companies = row.find_all('td')
    individual_row_data_largest_companies = [data.text.strip() for data in row_data_largest_companies]
    df_largest_companies.loc[len(df_largest_companies)] = individual_row_data_largest_companies

# Extract the data rows from the second table
data_rows_profitable_companies = table_profitable_companies.find_all('tr')[1:]
for row in data_rows_profitable_companies:
    row_data_profitable_companies = row.find_all('td')
    individual_row_data_profitable_companies = [data.text.strip() for data in row_data_profitable_companies]
    df_profitable_companies.loc[len(df_profitable_companies)] = individual_row_data_profitable_companies

# Print the DataFrame for the list of the largest companies
print(df_largest_companies)

# Print the DataFrame for the list of companies by profit
print(df_profitable_companies)

# Save the DataFrames to CSV files
df_largest_companies.to_csv('Largest Companies.csv', index=False)
df_profitable_companies.to_csv('Profitable Companies.csv', index=False)
```
